## Подготовка данных к когортному анализу

Подготовьте данные к когортному анализу с учётом того, что вы будете изучать окупаемость привлечения пользователей за первые 28 дней.

Для каждого пользователя найдите:
- дату первой сессии (дата привлечения пользователя, `first_dt`);
- канал привлечения (`first_channel`);
- стоимость привлечения пользователя (`cac`).

Результаты сохраните в новой таблице, например, `profiles`.

Добавьте к данным с событиями и покупками информацию о первом источнике привлечения и дату привлечения пользователя из созданной таблицы. После этого рассчитайте дневной лайфтайм для каждого события или покупки и добавьте в таблицу в виде целого числа. В дальнейшем при проведении когортного анализа используйте информацию о первом источнике привлечения пользователя.

```python
# Определяем первый месяц и канал активности для каждого пользователя
profiles = (
    df_visits
    .sort_values(['user_id', 'session_start']) 
    .groupby('user_id', as_index=False) 
    .agg({'session_start' : 'min', 'channel': 'first'}) 
    .rename(columns={'session_start': 'first_dt', 'channel': 'first_channel'})
)
```

```python
profiles.head()
```

```python
profiles['first_dt'] = profiles['first_dt'].dt.date
```

```python
# 2. CAC
user_counts = (
    profiles
    .groupby(['first_channel', 'first_dt'])['user_id'].nunique()
    .reset_index()
    .rename(columns={'user_id':'users_count'})
)
```

```python
user_counts.head()
```

```python
df_costs['dt'] = df_costs['dt'].dt.date
```

```python
costs = (
    df_costs
    .rename(columns={'dt': 'first_dt', 'channel': 'first_channel'})
    .groupby(['first_channel', 'first_dt'])['costs']
    .sum()
    .reset_index()
)
```

```python
cac_df = pd.merge(costs, user_counts, on=['first_channel', 'first_dt'], how='left')
```

```python
cac_df['cac'] = cac_df['costs'] / cac_df['users_count']
```

```python
cac_df.head()
```

```python
profiles = profiles.merge(cac_df[['first_channel', 'first_dt', 'cac']], on=['first_channel', 'first_dt'], how='left')
```

```python
profiles.head()
```

```python
profiles['cac'].sum()
```

```python
df_costs['costs'].sum()
```

```python
# 3. Добавляем инфо к событиям и покупкам
df_visits = df_visits.merge(profiles[['user_id', 'first_dt', 'first_channel']], on='user_id', how='left')
df_orders = df_orders.merge(profiles[['user_id', 'first_dt', 'first_channel']], on='user_id', how='left')
```

```python
df_visits['first_dt'] = pd.to_datetime(df_visits['first_dt'])
df_orders['first_dt'] = pd.to_datetime(df_orders['first_dt'])
```

```python
# 4. Дневной лайфтайм
df_visits['diff_day'] = (df_visits['session_start'] - df_visits['first_dt']).dt.days
df_orders['diff_day'] = (df_orders['event_dt'] - df_orders['first_dt']).dt.days
```

```python
# Выделяем из даты месяц активности 
df_visits['activity_month'] = df_visits['session_start'].dt.to_period('M').dt.to_timestamp()
```

```python
# Группировка для расчета MAU
mau_all = df_visits.groupby('activity_month')['user_id'].nunique().reset_index()
```

```python
# Визуализация
plt.subplots(figsize=(14,6))
plt.plot(mau_all['activity_month'], mau_all['user_id'], marker='o')
plt.title('MAU')
plt.xlabel('Даты')
plt.ylabel('Значения')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

```python
dau_all['activity_month'] = dau_all['activity_day'].dt.to_period('M').dt.to_timestamp()
```

```python
# Средний DAU по месяцам 
dau_monthly_average = dau_all.groupby('activity_month')['user_id'].mean().reset_index()
```

```python
# Готовим данные для расчета Stickiness
stickiness_table = pd.merge(mau_all, dau_monthly_average, on='activity_month', how='left')
```

```python
# Вычисляем Stickiness
stickiness_table['stick'] = 100 * (stickiness_table['user_id_y'] / stickiness_table['user_id_x'])
```

```python
plt.subplots(figsize=(14,6))
plt.plot(stickiness_table['activity_month'], stickiness_table['stick'], marker='o')
plt.title('Stickiness')
plt.xlabel('Даты')
plt.ylabel('Stickiness (DAU/MAU, %)')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

```python
# Выделяем первый месяц активности
profiles['first_month'] = pd.to_datetime(profiles['first_dt']).dt.to_period('M').dt.to_timestamp()
```

```python
# Считаем средний CAC по месяцам
monthly_cac_df = profiles.groupby('first_month').agg(
    total_users = ('user_id','nunique'),
    avg_cac = ('cac','mean')) \
    .reset_index() 
```

```python
# Визуализируем средний CAC по месяцам 
plt.subplots(figsize=(14,6))
plt.plot(monthly_cac_df['first_month'], monthly_cac_df['avg_cac'], marker='o')
plt.title('Средний CAC по месяцам')
plt.xlabel('Даты')
plt.ylabel('Значения')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

```python
# Выделяем первый месяц активности
df_orders['first_month'] = df_orders['first_dt'].dt.to_period('M').dt.to_timestamp()
```

```python
# Создаем df без ноябрьской когорты и лайфтайм 28 дней
orders_28d = df_orders[
    (df_orders['diff_day'] < 28) & 
    (df_orders['first_month'] < '2024-11-01')
].copy()
```

```python
orders_28d.shape[0]
```

```python
# revenue по месяцам 
month_revenue = orders_28d.groupby(['first_month'])['revenue'].sum().reset_index()
month_revenue = month_revenue.rename(columns={'revenue': 'revenue_moth'})
```

```python
month_users = profiles.groupby(['first_month'])['user_id'].nunique().reset_index()
month_users = month_users.rename(columns={'user_id': 'users_month'})
```

```python
df_LTV = month_revenue.merge(month_users,  on='first_month', how='left')
```

```python
# считаем LTV 
df_LTV['ltv'] = df_LTV['revenue_moth'] / df_LTV['users_month']
```

```python
df_LTV['first_month'] = pd.to_datetime(df_LTV['first_month']).dt.to_period('M').dt.to_timestamp()
```

```python
# Визуализируем средний LTV по месяцам

plt.subplots(figsize=(12,5))

plt.plot(df_LTV['first_month'], df_LTV['ltv'], marker='o')

plt.title('LTV за 28 дней на одного пользователя в разрезе месяца')
plt.xlabel('Месяц привлечения')
plt.ylabel('LTV')
plt.grid()
plt.show()
```

```python
df_LTV.head()
```

```python
cac_agg = profiles.groupby('first_month')['cac'].mean().reset_index().rename(columns={'cac': 'avg_cac'})
```

```python
df_ROI = df_LTV.merge(cac_agg, on='first_month', how='left')
```

```python
df_ROI.head(10)
```

```python
df_ROI['ROI'] = (df_ROI['ltv'] - df_ROI['avg_cac']) / df_ROI['avg_cac'] * 100
```

```python
df_ROI.head()
```

```python
plt.subplots(figsize=(14,6))

plt.plot(df_ROI['first_month'], df_ROI['ROI'])
    
plt.axhline(0, color='black', linestyle='--')

plt.title('ROI за 28 дней по месяцам')
plt.xlabel('Месяц привлечения')
plt.ylabel('ROI, % ')
plt.grid()
plt.show()
```

Запуск спортивного контента и изменение маркетинговой стратегии привели к резкому увеличению числа новых пользователей и росту расходов на рекламу. Рост расходов на привлечение новых пользователей (CAC) опережает рост дохода на пользователя (LTV), а часть каналов не окупает вложения даже за длительный период.


- Стремительный рост CAC, особенно в TipTop, снизил рентабельность маркетинга и увеличил срок окупаемости.

- Основная доля бюджета расходовалась на FaceBoom и TipTop, при этом FaceBoom не приносит возврата инвестиций.

#### Какие источники приносят прибыль, а какие нет:

Прибыль приносят: RocketSuperAds, TipTop (с отсрочкой окупаемости), MediaTornado.

Не приносят прибыль: FaceBoom (не окупается ни за 4, ни за 10 недель).

#### Рекомендации 

**Сфокусировать бюджеты на эффективных каналах:**
- Увеличить вложения в RocketSuperAds и поддерживать TipTop при строгом контроле расходов и мониторинге CAC.

- FaceBoom временно сократить или полностью приостановить вложения, провести аудит рекламных материалов и стратегий работы с этим каналом.

**Мониторить динамику CAC и LTV:**
- Еженедельно анализировать сроки окупаемости новых когорт. При росте CAC сразу корректировать рекламные активности и креативы.

**Тестировать новые сегменты и креативы**
- Для удержания рентабельности TipTop необходимо искать новые подходы к удержанию и монетизации аудитории: сегментация, персонализированные предложения, доработка контента.

**Оптимизировать бюджетные каналы**
- MediaTornado — отличный вариант для масштабирования дешёвого охвата, но не стоит ожидать от него высокой вовлечённости или роста LTV.

**Вести стратегический анализ окупаемости**
- Отслеживать не только 28-дневный период, но и более длинные горизонты окупаемости, чтобы своевременно выявлять негативные тренды.

#### Общий итог:
Procrastinate Pro+ обладает потенциалом для дальнейшего роста, но только при грамотной оптимизации маркетинговых расходов и ставке на каналы, которые обеспечивают и быстрое привлечение, и реальную окупаемость вложений. Главная задача — уходить от убыточных каналов, фокусироваться на удержании, вовлечённости и долгосрочной монетизации новых пользователей.
